```python
s='Nikhil Kumar'
bc=0
vc=0
cc=0
for i in s:
  if i==' ':
    bc=bc+1
  elif i in 'aeiouAEIOU':
    vc=vc+1
  else:
    cc=cc+1
print('Vowels:',vc)
print('Consonants:',cc)
print('Spaces:',bc)
```

```python
s='Today is 20-jan-2026 Tuesday'
bc=vc=sc=d=alp=0
for i in s:
  if i==' ':
    bc=bc+1
  elif i in 'aeiouAEIOU':
    vc=vc+1
  elif (i.isalpha()):
    alp=alp+1
  elif (i.isdigit()):
    d=d+1
  else:
    cc=cc+1
print('Vowels:',vc)
print('special character:',cc)
print('Spaces:',bc)
print('alpha:',alp)
print('digit:',d)
```
