```python
#Pre processing data = to check the data whether it is right or wrong
import pandas as pd
import numpy as np

data = pd.DataFrame({
  'Plot':['A','B','C','D','E'],
  'Yield':[2500,np.nan,2800,np.nan,2600],
  'Rainfall':[150,200,np.nan,180,170],
  'Temperature':[28,30,29,np.nan,27]
})
print('==Data with Missing Values===')
print(data)
#Detecting missing data
print('\n==Detecting Missing Values===')
print('is null')
print(data.isnull())
print(f'\nTotal missing per column')
print(data.isnull().sum())
```

```python
print('\n==Filling strategies')
#Fill with mean
data_mean=data.copy()
print(data_mean)
# Updated to match the current column name 'Yield'
data_mean['Yield']=data_mean['Yield'].fillna(data_mean['Yield'].mean())
print('\n','Fill with mean')
print(data_mean['Yield'])
```

```python
data_median = data.copy()
data_median['Rainfall'] = data_median['Rainfall'].fillna(data_median['Rainfall'].median())
# Fixed: Changed 'rainfall' to 'Rainfall' to match the column name
print(data_median['Rainfall'])
```

```python
data_ffill=data.copy()
print(data_ffill)
data_fill=data_ffill.fillna(method='ffill')
print(data_fill)
```

```python
#Dropping Missing data
print('Dropping missing row')
data_dropped=data.dropna()
print(data_dropped)
print(f'original rows:{len(data)},After dropping:{len(data_dropped)}')
```
