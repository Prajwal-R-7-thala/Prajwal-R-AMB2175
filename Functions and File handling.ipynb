```python
def max(num1,num2):
  if num1>num2:
    result=num1
  else:
    result=num2
  return result
z=max(3,6)
print(z)
```

```python
def max(x,y,z):
  if x>y and x>z:
    result=x
  elif y>x and y>z:
    result=y
  else:
    result=z
  return result
z=max(3,6,8)
print(z)
```
